```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import math
import quimb as qu
from pathlib import Path
from matplotlib.colors import LogNorm, Normalize
```

```python
# import test data
gname = 'cohVec_L_' + str(50) + \
    '_V_' + str(int(2*1e3)).zfill(4) + \
    '_dt_' + str(int(0.01*1e3)).zfill(4)

with h5py.File('./test_6.hdf5', 'r') as f:
    data_test_6 = f[gname+'/X'][()]
```

```python
# import test data
gname = 'cohVec_L_' + str(50) + \
    '_V_' + str(int(2*1e3)).zfill(4) + \
    '_dt_' + str(int(0.01*1e3)).zfill(4)

with h5py.File('./test_7.hdf5', 'r') as f:
    data_test_7 = f[gname+'/X'][()]
```

```python
t = np.arange(0, 10, 0.01)[:-1]

rows= 5
columns = 3
ylim = 0.5
dt = 0.01

names = ['I1X2', 'I1Y2', 'I1Z2', 'X1I2', 'X1X2', 'X1Y2', 'X1Z2', 'Y1I2', 'Y1X2', 'Y1Y2', 'Y1Z2', 'Z1I2', 'Z1X2', 'Z1Y2', 'Z1Z2']

fig, axs = plt.subplots(rows, columns, figsize=(15,15), dpi=80)
plt.setp(axs, xlim=(0,10), ylim=(-ylim, ylim))

for i in range(rows):
    for j in range(columns):
        #if i == rows-1 & j == columns-1:
        #    continue
        axs[i, j].plot(t, [data_test_6[k][columns*i+j] for k in range(len(t))], label='1e-6')
        axs[i, j].plot(t, [data_test_7[k][columns*i+j] for k in range(len(t))], label='1e-7')

        axs[i, j].grid()
        axs[i, j].set_title(names[columns*i+j], x=0.5, y=0.85)
axs[4, 2].grid()
#plt.legend()
plt.grid()

fig = plt.gcf()
plt.show()
```
